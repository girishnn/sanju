```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv('/content/mpg.csv')
df
```

```python
df.drop('car name',axis=1)
```

```python
df.isna().sum()
```

```python
x=df.weight
x
```

```python
y=df.mpg
y
```

```python
plt.scatter(x,y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df[['weight']],df[['mpg']])
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
x_pred=lr.predict(x_train)
x_pred
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
mse=mean_absolute_error(x_test,y_pred)
mae=mean_squared_error(x_test,y_pred)
r2=r2_score(x_test,y_pred)
```

```python
rmse=np.sqrt(mse)
rmse
```

```python
print(mse)
print(mae)
print(r2)
```
